# **IMPORTING THE REQUIRED LIBRARIES**

```python
from bs4 import BeautifulSoup
import requests
import csv
import pandas as pd
import numpy as np
import html
```

## **FETCHING URL TO SCRAP FROM WIKIPEDIA**

```python
url = 'https://en.wikipedia.org/wiki/List_of_largest_companies_in_the_United_States_by_revenue'
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html')
```

Pulling information from soup

```python
print(soup)
```

```python
soup.find('table')
```

```python
soup.find_all('table')[0]
```

```python
soup.find_all('table', class_= 'wikitable sortable')
```

```python
table = soup.find_all('table')[1]
print(table)
```

```python
# viewing the "th" tags
world_titles = table.find_all('th')
print(world_titles)
```

```python
world_titles = [tittle.text.strip() for tittle in world_titles]
print(world_titles)
```

```python
df = pd.DataFrame(columns = world_titles)
df
```

```python
column_data = table.find_all('tr')
```

```python
for row in column_data [1:]:
  row_data = row.find_all('td')
  individual_row_data = [data.text.strip() for data in row_data]
  length = len(df)
  df.loc[length] = individual_row_data

df.head()
```

```python
df.tail()
```

```python
# exporting the data to a csv file
df.to_csv(r'LARGEST USA COMPANIES BY REVENUE.CSV', index=False)
```
